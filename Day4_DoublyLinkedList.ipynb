```python
class Node:

    def __init__(self,prev = None, item = None, next = None):
        self.prev = prev
        self.item = item
        self.next = next
```

```python
class DLLIterator:
    def __init__(self, start):
        self.current = start

    def __iter__(self):   # creating iterator
        return self

    def __next__(self):
        if not self.current:
            raise StopIteration
        data = self.current.item
        self.current = self.current.next
        return data
        
```
