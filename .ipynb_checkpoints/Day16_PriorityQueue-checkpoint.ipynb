### Priority Queue

```python
class PriorityQueue:
    def __init__(self):
        self.pQueue = []

    def is_empty(self):
        return len(self.pQueue) == 0

    def push(self,data,priority):
        index = 0  # lower number , higher priority
        while index < len(self.pQueue) and self.pQueue[index][1] <= priority:
            index += 1
        self.pQueue.insert(index, (data,priority))

    def pop(self):
        if self.is_empty() :
            raise IndexError("Priority Queue is empty")
        else:
            return self.pQueue.pop(0)[0]

    def size(self):
        return len(self.pQueue)        
```

```python
pq = PriorityQueue()
pq.push("Amit",4)
pq.push("bla bla",2)
pq.push("Yukti",7)
pq.push(7,9)
print(pq.size())
print(pq.pop())
print(pq.size())

print("--------")
i=0
while not pq.is_empty():
    print(i , " " ,pq.pop(), end = ' ')
    i += 1
```
