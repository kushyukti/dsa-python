### Recursion

Function calling itself is called recursion



```python
# sum of n natural numbers
def f1(n):
    if n == 1:
        print(n)
        return 1
    print(n,end="+")
    s = n + f1(n-1)
    return s

r = f1(5)
print("\n",r)
```

#### write a program to print first n natural numbers in reverse order

```python
def revPrint(n):
    if n == 0:
        return 0
    print(n, end = " ")

    return revPrint(n-1)

revPrint(7)
```

write a recursive program to print first n odd natural numbers in reverse order

```python
def oddNatural(n):
    if n == 0:
        return 1
    if n % 2 == 1:
        print(n, end=" ")
    return oddNatural(n-1)

oddNatural(7)
```

```python
def printNOdd(n):
    if n > 0:
        printNOdd(n-1)
        print(2 * n - 1, end= ' ')
printNOdd(10)
```

write a recursive program to print first n even natural numbers in reverse order

```python
def evenNatural(n):
    if n == 2:
        print(n, end = " ")
        return 2
    if n % 2 == 0:
        print(n, end = " ")
    return evenNatural(n-1)


evenNatural(6)
```

```python
def printNEven(n):
    if n > 0:
        printNOdd(n-1)
        print(2 * n , end= ' ')
printNEven(10)
```

```python
def printNOddReverse(n):
    if n > 0:
        print('\n',2 * n - 1, end= ' ')
        printNOddReverse(n-1)
printNOddReverse(10)
```

write a recursive program to print first n odd natural numbers in reverse order

```python
def oddReverse(n):
    if n == 1:
        print(n , end=" ")
        return 1
    if n % 2 == 1:
        print(n , end=" ")
    return oddReverse(n-1)

oddReverse(7)
```
